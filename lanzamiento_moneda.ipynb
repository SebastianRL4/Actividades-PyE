```python
#Sebastian Rodriguez Lopez
#Materia:Probabilidad y Estadística
#Simulación de lanzamientos de moneda

import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
np.random.seed(42)

#variables
n=10 #numero de lanzamientos de la moneda en cada experimento
p=0.5 #tomamos para este caso la probabilidad de obtener cara
size=10000 #numero de veces que se repite el experimento
x=[np.random.binomial(n=n, p=p, size=size)] #obtenemos el número de caras de los 10 lanzamientos para los 10000 experimentos. Lo que resulta en 100000 lanzamientos
a=[np.equal(x,i).sum() for i in range(n)] #obtenemos el numero de caras de cada 10 lanzamientos repitiendo el experimento 10000 veces
prob=[np.equal(x,i).mean() for i in range(n)] #calculamos la probabilidad de obtener x caras en n=10
#resultados
print(x)
print(a)
print(prob)
plt.xticks(range(n))
plt.plot(list(range(n)), prob, color='blue', marker='o')
plt.xlabel('Numero de caras', fontsize=14)
plt.ylabel('Probabilidad', fontsize=14)
```
